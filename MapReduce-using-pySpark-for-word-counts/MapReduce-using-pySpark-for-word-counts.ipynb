```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, split, lower, trim

# Create a SparkSession
spark = SparkSession.builder \
    .appName("WordCount") \
    .getOrCreate()

# Load the text file as a DataFrame
text_df = spark.read.text("/Users/eshitashitij/Downloads/shakepoems.txt")

# Split each line into words and convert to lowercase
word_pairs = text_df.select(explode(split(lower(trim(text_df.value)), ' ')).alias('word'))

# Map each word to a tuple (word, 1)
word_counts_tuples = word_pairs.rdd.map(lambda row: (row.word, 1))

# Reduce by key (word) by summing up the counts
word_counts_reduced = word_counts_tuples.reduceByKey(lambda a, b: a + b)

# Sort the word counts by count (ascending order)
word_counts_sorted = word_counts_reduced.sortBy(lambda x: x[1], ascending=False)

# Print the sorted word counts
for word, count in word_counts_sorted.collect():
    print(f"{word}: {count}")

# Save the word counts to a file
word_counts_sorted.saveAsTextFile("/Users/eshitashitij/Downloads/poem_wc.txt")

# Stop the SparkSession
#spark.stop()
```
